```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
from pytrends.request import TrendReq
```

# Setup pytrend library and defining keyword

```python
pytrends = TrendReq(hl = 'en-US', tz=360)
keyword = "cricket"
```

# Requesting Data

```python
pytrends.build_payload([keyword], cat=0,timeframe = 'today 12-m', geo='', gprop='')
```

# Interest Country wise

```python
data_region = pytrends.interest_by_region()
data_region = data_region.sort_values(by = keyword,ascending = False).head(20)
```

```python
plt.figure(figsize = (10,6))
sns.barplot(x = data_region[keyword], y = data_region.index, palette = "Blues_d")
plt.title(f"'{keyword}' search by Top Countries")
plt.xlabel("Interest")
plt.ylabel("Countries")
plt.show()
```

# Showing World Map

```python
data_region = data_region.reset_index()
fig = px.choropleth(data_region,
                   locations = 'geoName',
                   locationmode = 'country names',
                   color = keyword,
                   title = f"Searching Interest for '{keyword}' by Country",
                   color_continuous_scale = 'Blues')
fig.show()
```

# time wise interest

```python
time_df = pytrends.interest_over_time()
```

```python
plt.figure(figsize=(12,6))
plt.plot(time_df.index, time_df[keyword], marker = 'o', color = 'red')
plt.title(f"Search interest over time'{keyword}'")
plt.xlabel("Date")
plt.ylabel("Interest")
plt.grid(True)
plt.show()
```

# Multiple Keyword comparison

```python
list_key = ["cricket","foot ball", "basket ball"]
pytrends.build_payload(list_key, cat=0, timeframe = 'today 12-m', geo='',gprop='')
```

```python
compare_df = pytrends.interest_over_time()

plt.figure(figsize=(12,6))
for key in list_key:
    plt.plot(compare_df.index, compare_df[kw], label=key)

plt.title("Comparison of keywords over time")
plt.xlabel("Date")
plt.ylabel("Interest")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
