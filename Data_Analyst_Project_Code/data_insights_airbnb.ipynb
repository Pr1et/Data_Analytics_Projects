```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Airbnb_dataset.csv')
```

```python
df.head()
```

```python
df.columns
```

# Checking the missing values

```python
print(df.isnull().sum())
```

```python
df.info()
```

# Correcting the data types of columns

```python
df['last review'] = pd.to_datetime(df['last review'], errors='coerce')

df.fillna({'reviews per month': 0, 'last review': df['last review'].min()}, inplace=True)

df.dropna(subset=['NAME','host name'], inplace=True)
```

```python
print(df.isnull().sum())
```

```python
df.info()
```

```python
df['price'] = df['price'].replace('[\$,]', '', regex=True).astype(float)
df['service fee'] = df['service fee'].replace('[\$]', '', regex=True).astype(float)
```

# Removing the Duplicate Value

```python
df.drop_duplicates(inplace=True)
```

```python
print(df.info())
```

# Removing the columns which has most null values and not necessary in dataset

```python
df = df.drop(columns=['house_rules','license'], errors='ignore')
```

```python
df.info()
```

# Descriptive Statistics

```python
df.describe()
```

# Data Visualization

# Distribution of prices

```python
plt.figure(figsize=(12,6))
sns.histplot(df['price'],bins=50, kde=True, color='Green')
plt.title('Listing Price Distribution')
plt.xlabel('Prices in $')
plt.ylabel('Count Frequency')
plt.show()
```

# Analysis of Room types

```python
plt.figure(figsize=(12,6))
sns.countplot(x='room type',data=df, color='lightblue')
plt.title('Distribution of Room Types')
plt.xlabel('Room Types')
plt.ylabel('Count')
plt.show()
```

# Analysis of Neighborhood

```python
plt.figure(figsize=(12,6))
sns.countplot(y='neighbourhood group',data=df, color='red', order=df['neighbourhood group'].value_counts().index)
plt.title('Numbers of Listing Neighborhoods')
plt.xlabel('Counts')
plt.ylabel('Neighborhoods')
plt.show()
```

# Comparison of Prices and Room Types

```python
plt.figure(figsize=(12,6))
sns.boxplot(x='room type',y='price',data=df, hue='room type',palette='Set2')
plt.title('Box Plot of Room Type V/S Prices')
plt.xlabel('Room Types')
plt.ylabel('Prices in $')
plt.legend(title='Room Types')
plt.show()
```

# Reviews OverTime

```python
Review_overtime = df.groupby(df['last review'].dt.to_period('M')).size()

plt.figure(figsize=(12,6))
Review_overtime.plot(kind='line', color='Purple')
plt.title('Distribution of Reviews OverTime')
plt.xlabel('Date')
plt.ylabel('Total Reviews')
plt.show()
```
