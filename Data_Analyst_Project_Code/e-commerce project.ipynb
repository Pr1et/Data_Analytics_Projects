```python
import pandas as pd
import plotly.express as px
import plotly.graph_objects as go
import plotly.io as pio
import plotly.colors as colors
from seaborn import color_palette

pio.templates.default = "plotly_white"
```

```python
df = pd.read_csv('Superstore_dataset.csv', encoding='latin1')
```

```python
df.head()
```

```python
df.describe()
```

```python
df.info()
```

order date and ship data is in object type, so we have to convert it in date datatype

```python
df['Order Date'] = pd.to_datetime(df['Order Date'])
df['Ship Date'] = pd.to_datetime(df['Ship Date'])
```

```python
df.info()
```

```python
df['Order Month']= df['Order Date'].dt.month
df['Order Year'] = df['Order Date'].dt.year
df['Order Day of Week'] = df['Order Date'].dt.dayofweek
```

```python
df.head()
```

# Sales Analysis by Months

```python
sales_by_months = df.groupby('Order Month')['Sales'].sum().reset_index()
fig = px.line(sales_by_months, x='Order Month', y='Sales', title='Sales by Month')
fig.show()
```

# Sales by Category

```python
sales_by_category = df.groupby('Category')['Sales'].sum().reset_index()
fig = px.pie(sales_by_category, values='Sales', names='Category', hole=0.1, color_discrete_sequence=colors.qualitative.Pastel2)

fig.update_traces(textposition = 'inside', textinfo='percent+label')
fig.update_layout(title_text='Sales by Category', title_font=dict(size=14))
fig.show()
```

# Sales by Sub Category

```python
sales_by_subcategory = df.groupby('Sub-Category')['Sales'].sum().reset_index()

fig = px.bar(sales_by_subcategory, x='Sub-Category', y='Sales', title='Sales by Sub-Category')

fig.show()
```

# Profit Analysis by Months

```python
profit_by_months= df.groupby('Order Month')['Profit'].sum().reset_index()

fig = px.line(profit_by_months, x='Order Month', y ='Profit', title='Profit by Month')
fig.show()
```

# Profit by Category

```python
profit_by_category = df.groupby('Category')['Profit'].sum().reset_index()

fig = px.bar(profit_by_category, x='Category', y='Profit', title='Profit by Category')
fig.show()
```

# Profit by Sub Category

```python
profit_by_subcategory = df.groupby('Sub-Category')['Profit'].sum().reset_index()

fig = px.bar(profit_by_subcategory, x='Sub-Category', y='Profit', title='Profit by Sub-Category')
fig.show()
```

# Customer Segment (Sales & Profit)

```python
sales_profit_segment = df.groupby('Segment').agg({'Sales': 'sum', 'Profit': 'sum'}).reset_index()

color_palette = colors.qualitative.Pastel2

fig = go.Figure()
fig.add_trace(go.Bar(x=sales_profit_segment['Segment'],
                     y=sales_profit_segment['Sales'],
                     name='Sales',
                     marker_color=color_palette[0]))

fig.add_trace(go.Bar(x=sales_profit_segment['Segment'],
                     y=sales_profit_segment['Profit'],
                     name='Profit',
                     marker_color=color_palette[1]))

fig.update_layout(title='Sales & Profit by Segment',
                  xaxis_title='Customer Segment',yaxis_title='Amount')
fig.show()
```

```python
sales_profit_segment = df.groupby('Segment').agg({'Sales': 'sum', 'Profit':'sum'}).reset_index()
sales_profit_segment['Sales_to_Profit_Ratio'] = sales_profit_segment['Sales']/sales_profit_segment['Profit']

print(sales_profit_segment[['Segment', 'Sales_to_Profit_Ratio']])
```
